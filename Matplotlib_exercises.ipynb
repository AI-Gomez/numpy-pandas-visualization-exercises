```python
import matplotlib.pyplot as plt
```

```python
#Use matplotlib to plot the following equation: y=x^2-x+2
#You'll need to write the code that generates the x and y points.
#Add an anotation for the point 0, 0, the origin.
x = range(1,50)
def equation(n):
    return n**2 - n + 2
y = [equation(n) for n in x]

plt.plot(x,y)
plt.annotate('the origin', xy=(0, 100))
```

```python
#Create and label 4 separate charts for the following equations 
#(choose a range for x that makes sense):
#y=squareroot(x)

import math
x = range(1,50)
def square(n):
    return math.sqrt(n)
y = [square(n) for n in x]
plt.plot(x,y)
plt.annotate('the origin', xy=(3, 1))
```

```python
#y=x^3
x = range(1,50)
def cubed(n):
    return n**3
y = [cubed(n) for n in x]

plt.title('Cubed')
plt.plot(x,y)
plt.annotate('the origin', xy=(1, 2500))
```

```python
#y=2^x
x = range(1, 10)
def nsquare(n):
    return 2**n
y = [nsquare(n) for n in x]

plt.plot(x,y)
plt.annotate('the origin', xy=(1, 20))
```

```python
#3 Combine the figures you created in the last step into one large 
#figure with 4 subplots.
x = range(1,10)
def equation(n):
    return n**2 - n + 2
y0 = [equation(n) for n in x]


plt.subplot(2,2,1)
plt.plot(x,y0)
plt.annotate('the origin', xy=(0, 100))
#####
import math
x = range(1,10)
def square(n):
    return math.sqrt(n)
y1 = [square(n) for n in x]


plt.subplot(2,2,2)
plt.plot(x,y1)
plt.annotate('the origin', xy=(3, 1))
#####
x = range(1,10)
def cubed(n):
    return n**3
y2 = [cubed(n) for n in x]


plt.subplot(2,2,3)
plt.plot(x,y2)
plt.annotate('the origin', xy=(1, 2500))
#####
x = range(1, 10)
def nsquare(n):
    return 2**n
y3 = [nsquare(n) for n in x]


plt.subplot(2,2,4)
plt.plot(x,y3)
plt.annotate('the origin', xy=(1, 20))
```

```python
#Combine the figures you created in the last step into one figure where 
#each of the 4 equations has a different color for the points. Be sure 
#to include a legend.
x = range(1,10)

y0 = [equation(n) for n in x]
y1 = [square(n) for n in x]
y2 = [cubed(n) for n in x]
y3 = [nsquare(n) for n in x]

plt.plot(x,y0, label = 'X^2-X+2')
plt.plot(x,y1, label = 'sqrtX')
plt.plot(x,y2, label = 'X^3')
plt.plot(x,y3, label = '2^X')

plt.legend()
```
