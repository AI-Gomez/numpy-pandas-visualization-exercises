```python
import matplotlib.pyplot as plt
import math
```

```python
x = range(1,4)
y0 = [n*0+1 for n in x]
y1 = [math.log(n) for n in x]
y2 = [n for n in x]
y3 = [n*math.log(n) for n in x]
y4 = [n**2 for n in x]
y5 = [n**n for n in x]

def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
y6 = [factorial(n) for n in x]

y7 = [n**n for n in x]

plt.figure(figsize = (16, 9))

plt.plot(x,y0, label = '0(1)')
plt.plot(x,y1, label = '0(log n)')
plt.plot(x,y2, label = '0(n)')
plt.plot(x,y3, label = '0 (n log n)')
plt.plot(x, y4, label = '0(n^2)')
plt.plot(x, y5, label = '0(2^n)')
plt.plot(x, y6, label = '0(n!)')
plt.plot(x, y7, label = '0(n^n)')

plt.title('Big O Notation')
plt.xlabel('Operations')
plt.ylabel('Elements')

plt.legend()
plt.show()
```
